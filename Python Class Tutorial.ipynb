A method may take arguments, it defined that way

```python
class fruit:
    """
    This class creates instances of fruits
    """
    def size(self, x):
        print(f"I am size {x}")
```

```python
orange=fruit()
orange.size(7)
```

A python class may also have some special attributes, like __doc__ for the docstring;

```python
fruit.__doc__
```

we will take an example including the __init__() magic method and self parameter

```python
class fruit:
    """
    This class creates instances of fruits
    """
    def __init__(self, color, size):
        self.color=color
        self.size=size
    def salutation(self):
        print(f"I am {self.color} and a size {self.size}")
```

```python
orange=fruit("Orange",7)
```

```python
orange.salutation()
```

As we can see, the __init__() method is equivalent to a constructor in c++ or java. It gets called everytime we create an object of the class. Likewise, the self parameter is to tell the interpreter to deal with the current object. This is like the 'this' keyword in Java. But we dont have to call it 'self' . We can call it anything The object is passed as the first argument, fitting in for 'self'. 

```python
fruit.salutation
```

```python
orange.salutation
```

```python
sayhi = orange.salutation
```

## Attributes belonging to Python class

For explaining this, we will redefine the python class fruit

```python
class fruit:
    size='Small'
    def __init__(self, color, shape):
        self.color=color
        self.shape=shape
    def salutation(self):
        print(f"I am {self.color} and a shape {self.shape}")
```

```python
fruit.size
```

```python
orange=fruit('orange','round')
```

```python
orange.size
```

Here, the attribute 'size' belongs to the class, but we can call it on an object as well
