### Slicing a Tuple in Python

```python
percentages=(99,95,90,89,93,96)
```

```python
percentages[2:4]
```

```python
percentages[:4]
```

```python
percentages[4:]
```

```python
percentages[2:2]
```

```python
# Unlike positive indexing, it begins traversing from the right.
percentages[:-2]
```

```python
# This prints out the items from the tuples beginning to two items from the end.
percentages[-2:]
```
