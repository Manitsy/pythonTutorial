## Slicing a Python list

#### when you want only a part of a python list, you can use the slicing operator[].

```python
indices = ['Zero','one','two', 'three', 'four', 'five']
```

```python
indices[2:4]
```

#### This retuns items from index 2 to index 4-1 (i,e 3)

```python
# this return item from the beginning of the list to index 3.
indices[:4]
```

```python
# it will returns items from index 4 to the end of the list in python.

indices[4:]
```

```python
# it returns the whole list

indices[:]
```

```python
indices
```

### Negative indices:

#### The indices we mention can be negative as well. A negative index mean travesal from the end of the list.

```python
indices[:-2]

# this returns item from the list's beginning to two item from the end.
```

```python
indices[1:-2]

# It returns items from the item at index 1 to two items from the end.
```

```python
indices[-2:-1]

# This returns items from two from the end to one from the end.
```

```python
indices[-1:-2]
# This returns an empty Python list, because the start is ahead of the stop for the traversal.
```

## Python List Comprehension

#### You can create a new list just like you would do in mathematics. To do so, type an expression followed by a for statement


```python
even = [2*i for i in range(1,11)]
```

```python
even
```

```python
even1 = [
        for i in range(1,11):
            i*2
        ]
```

```python
for i in range (1,11):
    i*2
```

#### Optionally, you can add an if-statement to filter out item. If we want to change the list to hold only those items from 1 to 20 that are even and are divisible by 3,

```python
even = [2*i for i in range(1,11) if i%3 == 0]
```

```python
even
```
