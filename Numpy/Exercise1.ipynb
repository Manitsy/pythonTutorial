Calculate total amount earn by end of Thrum

```python
import numpy as np
```

```python
cost_of_item= np.array([3,4,2])
```

```python
cost_of_item
```

```python
no_of_item_sold = np.array([[13,9,7,15],
                            [8,7,4,6],
                            [6,4,0,3],
                            ])
```

```python
no_of_item_sold
```

```python
amount_earn=np.dot(cost_of_item,no_of_item_sold)
```

```python
amount_earn
```

```python
total_amount_earn=np.sum(amount_earn)
```

```python
total_amount_earn
```

### When we do multiplication

#### 1: The number of columns of the 1st matrix must equal the number of rows of the 2nd matrix
  <tab>**no. of column of 1st matrix=no.of rows of 2nd matrix**</tab>

#### 2: The result will have the same number of rows as the 1st matrix, and the same number of column as the 2nd column
