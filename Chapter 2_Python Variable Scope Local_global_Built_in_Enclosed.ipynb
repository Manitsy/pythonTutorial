## 2. What is Python Variable scope?

### The scope of a variable in python is that part of the code where it is visible. Actually, to refer to it, you dont need to use any prefixes then.

### Let's take an example, but before let's revise python syntax

```python
b=8
```

```python
def func():
    a=7
    print(a)
    print(b)
```

```python
func()
```

```python
a
```

```python
b
```

#### Also, the duration for which a variable is alive is called its 'lifetime'

### 3. Types of Python Variable scope.

#### There are 4 types of variable scope in python, let's discuss them one by one:

### a. Local Scope

#### In the above code, we define a variable "a" in a function 'func'. So, 'a' is local to 'func'. Hence, we can read/write it in func, but not outside it. When we try to do so, it raises a NameError.

#### Look at this code.

```python
a=0
```

```python
def func():
    print(a)
    a=1
    print(a)
```

```python
func()
```

#### Hence, we could have accessed the global scope 'a' inside 'func', but since we also define a local 'a' in it, it no longer accesses the global 'a'. Then the first print statement inside 'func' throws an error in python.
